```python
#before running the script, import your soundfile and rename it output.wav
import crepe
from scipy.io import wavfile
sr, audio = wavfile.read('output.wav')
time, frequency, confidence, activation = crepe.predict(audio, sr, viterbi=True)
```

```python
print(len(confidence))
```

```python
notes=["begin","o","o","o"]

for i in range(300):
    if confidence[i]>0.8:
        x=int(frequency[i])
        #print(x)
        if x>970 and x<1010:
            if notes[1]=="o":
                notes[1]="B5"
            elif notes[1]=="B5":
                notes[1]="B5"
            elif notes[2]=="o":
                notes[2]="B5"
            elif notes[2]=="B5":
                notes[2]="B5"
            elif notes[3]=="o":
                notes[3]="B5"
            elif notes[3]=="B5":
                notes[3]="B5"
            else:
                print("B5")
        if x>1020 and x<1070:
            if notes[1]=="o":
                notes[1]="C6"
            elif notes[1]=="C6":
                notes[1]="c6"
            elif notes[2]=="o":
                notes[2]="C6"
            elif notes[2]=="C6":
                notes[2]="C6"
            elif notes[3]=="o":
                notes[3]="C6"
            elif notes[3]=="C6":
                notes[3]="C6"
            else:
                print("C6")
        if x>1630 and x<1690:
            if notes[1]=="o":
                notes[1]="G#6"
            elif notes[1]=="G#6":
                notes[1]="G#6"
            elif notes[2]=="o":
                notes[2]="G#6"
            elif notes[2]=="G#6":
                notes[2]="G#6"
            elif notes[3]=="o":
                notes[3]="G#6"
            elif notes[3]=="G#6":
                notes[3]="G#6"
            else:
                print("G#6")
print(notes)
```
